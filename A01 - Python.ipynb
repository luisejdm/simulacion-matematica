## Punto 3

```python
tabla2 = pd.read_excel("datosCT.xlsx")

tabla2
```

### Utilidades

```python
tabla2["Utilidad"]=tabla2["Ingreso"]-tabla2["Gasto"]

tabla2
```

### Promedio de gastos cuando la utilidad fue negativa

```python
mper = tabla2[tabla2["Utilidad"]<0]

mper
```

```python
promedio = mper["Gasto"].mean() #el .mean saca el promedio de la columna seleccionada

print(promedio)
```

### Ingresos totales

```python
totalingresos=tabla2["Ingreso"].sum() # el .sum suma toda la columna seleccionada

print(totalingresos)
```
