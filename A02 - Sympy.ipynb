**Para la siguientes funciones encuentre los puntos críticos, grafique y diga si éste se trata de un máximo o mínimo, local o global.**

$f(x)=1260+117x-9x^2$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


sp.var("x")


# 1. Definir función
def f(x):
    return 1260 + 117*x - 9*x**2

#2 2. Derivar
df = sp.diff(f(x), x)

# 3. Resolvemos igualando a cero
pc = sp.solve(df, x)

# 4. Segunda derivada
ddf = sp.diff(df, x)

# 5. Graficar
xn = np.linspace(-7,20,1000)

plt.figure(figsize=(4,4))
plt.plot(xn, f(xn))

# 6. Graficar puntos críticos
for c in pc:
    plt.plot(c, f(c), "*r", ms=10)
    
# 7. Imprimir si es máximo o mínimo.

for c in pc:
    crit_2da_derivada = ddf.subs(x, pc)
    if crit_2da_derivada < 0:
        print("El punto crítico se encuentra en " + str(pc[0]) + " y es un máximo.")
    elif crit_2da_derivada > 0:
        print("El punto crítico se encuentra en " + str(pc[0]) + " y es un mínimo.")
    else:
        print("El punto crítico se encuentra en " +str(pc[0]) + " y no se puede definir por el criterio de la segunda derivada")
    
```

#### Debido a que la función es parabólica, el punto crítico es un máximo global de la función.

**1) El perímetro de un triángulo isósceles es de $10 cm$. ¿Cuánto deben medir sus lados para que el volumen del cuerpo generado por la rotación del triángulo en torno a su base sea el mayor posible? (Volumen de un cono $= \frac{1}{3}\pi r^2 h$, donde $r$ es el radio de la base y $h$ la altura del cono).**

Debido a que es un triángulo isósceles, se tienen 2 lados iguales y una base, que en este caso partirémos en 2 para encontrar el radio mas fácil. La ecuación nos quedaría de la siguiente manera.

$ 2x+2y=10 $

Sin embargo, la altura en el triángulo es importante, así como el radio, que sería la mitad de la base del tríangulo, así que podemos separar el triángulo isósceles en 2 triángulos rectángulos. La ecuación nos quedaría de la manera:

$x+y=5$

Ademas en la fórmula del volumen del cono hay que usar la altura del triángulo, por lo que se puede aplicar el teorema de pitágoras:

$ h = \sqrt{y^2 - x^2} $

Una vez que se tienen estas fórmulas base, podemos sustituirlas en la fórmula del volumen del cono, para despues pasar al proceso de optimización. Las variables que reemplazaré en la fórmula del cono será $h$ y $y$. Pero antes se tiene que despejar $y$ de la fórmula de los lados del triángulo y quedaría de la siguiente manera:

```python
import sympy as sp

sp.var("x")
sp.var("y")

ecuacion_triangulo = x + y -5
eq_isosceles = 2*x + 2*y - 10

y_despejada = sp.solve(ecuacion_triangulo,y)

y_despejada
```

Ahora podemos remplazar el radio $r$ por $x$, $y$ por $5-x$ y $h$ por $\sqrt{y^2 - x^2}$ de tal manera que nos quede una equación con sólo una variable.

$ V = \frac{𝜋}{3}x^2 \sqrt{(5-x)^2-(x^2)} $

Una vez que se tiene la ecuación con solo la variable $x$ podemos iniciar el proceso de optimización.

1. Primero definir la función

```python
def f(x):
    return (sp.pi/3) * x**2 * sp.sqrt((5-x)**2 - x**2)
f(x)
```

2. Derivarla

```python
df = sp.diff(f(x),x)
df
```

3. Resolvemos igualando a cero

```python
pc = sp.solve(df, x)
pc
```

4. Calcular la segunda derivada

```python
ddf = sp.diff(df, x)
ddf
```

5. Criterio de la segunda derivada

```python
for c in pc:
    crit_2da_derivada = ddf.subs(x, c)    
    if crit_2da_derivada < 0:
        print("El punto crítico se encuentra en " + str(c) + " y es un máximo.")
    elif crit_2da_derivada > 0:
        print("El punto crítico se encuentra en " + str(c) + " y es un mínimo.")
    else:
        print("El punto crítico se encuentra en " +str(c) + " y no se puede definir por el criterio de la segunda derivada")
    
```

**2) Disponemos de una barra de aluminio de 6 metros para construir una portería de fútbol. Si queremos que el área de la portería sea máxima, ¿cuánto deben medir los postes y el travesaño?**

Para construir una portería se necesitan dos postes que miden lo mismo y un travesaño en la parte superior. Con ello podemos formular la siguiente ecuación:

$2x + y = 6$

Para maximizar el area primero hay que recordar que el area de un cuadrilátero es:

$A=b*h$

La altura de $h$ de la portería sería uno de los dos postes iguales, por lo que podríamos representar $h$ como $x$ de la siguiente manera:

$A=b*x$

Aunque en la base no se pondrá un poste de aluminio, si sabemos que mide lo mismo que el travesaño. Así que podemos representar $b$ como $y$ de la siguiente manera:

$A=y*x$

Sin embargo, no se puede derivar una función con dos variables, por lo que tenemos que representar  $y$ de otra manera usando $x$. Así que la despejaremos de la primera ecuación en donde se tienen las medidas de las barras de aluminio.

```python
import sympy as sp

sp.var("x")
sp.var("y")

eq_barras = 2*x + y - 6

y_despejada = sp.solve(eq_barras,y)
y_despejada
```

Ahora podemos reemplazar el valor anterior en la ecuación del area de la portería de la siguiente manera:

$A=(6-2x)*x=6x-2x^2$

Una vez que se tiene la ecuación con solo la variable $x$ podemos iniciar el proceso de optimización.

1. Definir la función.

```python
def f(x):
    return 6*x - 2*x**2
f(x)
```

2. Derivar la función.

```python
df = sp.diff(f(x),x)
df
```

3. Resolver Igualando a 0.

```python
pc = sp.solve(df, x)
pc
```

4. Calcular la segunda derivada.

```python
ddf = sp.diff(df, x)
ddf
```

5. Criterio de la segunda derivada.

```python
for c in pc:
    crit_2da_derivada = ddf.subs(x, c)    
    if crit_2da_derivada < 0:
        print("El punto crítico se encuentra en " + str(c) + " y es un máximo.")
    elif crit_2da_derivada > 0:
        print("El punto crítico se encuentra en " + str(c) + " y es un mínimo.")
    else:
        print("El punto crítico se encuentra en " +str(c) + " y no se puede definir por el criterio de la segunda derivada")
    
```

Por lo tanto si $\frac{3}{2}$ es un máximo, solo basta con sustituir en las ecuaciones anteriores para conocer las medidas de los postes que maximizan el area de la portería.

Sustituyendo en la ecuación de las medidas de los postes:

```python
sp.solve(eq_barras.subs(x,3/2),y)
```
